````python
from visify import visualize
````

## 1. Neural network — describe architecture

````python
# Describe any architecture with a layers list
# Click Next / Auto to walk through the forward pass
visualize(model='neural_network', layers=[4, 8, 8, 2])
````

## 2. Neural network — with real input (shows activations)

````python
import numpy as np

x = np.array([0.8, 0.2, 0.6, 0.9])   # your actual input
visualize(model='neural_network', layers=[4, 8, 8, 2], input=x)
````

## 3. Neural network — deep architecture

````python
visualize(
    model='neural_network',
    layers=[8, 16, 16, 8, 4, 2],
    activation='relu'
)
````

## Plugin API — register your own visualizer
```python
from visify import Registry
from visify.core.base import BaseVisualizer, Frame

class MyVisualizer(BaseVisualizer):
    def frames(self):
        yield Frame(type='custom', label='Hello!', data=self.value)

Registry.register('my_thing', MyVisualizer)
visualize(my_thing='some_value')
```
